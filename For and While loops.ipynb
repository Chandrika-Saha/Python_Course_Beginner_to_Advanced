# Loops and iterations - For/ While loops

```python
nums = [1, 2, 3, 4, 5]
print(nums)

for num in nums:
    # num = 1
    # num = 2
    # num = "Hello" ...
    print(num)
```

```python
nums = [1, 2, 3, 4, 5, 4, 6]
# find the number 4
for num in nums:
    if(num == 4):
        print("Found 4!")
        break
    print(num)
```

```python
# find the number 4
nums = [1, 2, 3, 4, 5, 4, 6]
for num in nums:
    if(num == 4):
        print("Found 4!")
        continue
    print(num)
```

```python
nums = [1, 2, 3]
for num in nums:
    # num = 1
    # num = 2
    for letter in "ab":
        # letter = "a"
        # letter = "b"
        print(num, letter)
```
